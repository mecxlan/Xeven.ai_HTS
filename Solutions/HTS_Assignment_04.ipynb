```python
"""Write a function called calculate_sum that takes a list of numbers as an
argument and returns the sum of those numbers"""

#defining function to calculate sum of a given list
def calculate_sum(numbers_list):
  #defining a variable to store sum
  sum = 0
  #iterating for loop for all numbers in given list
  for i in range(len(numbers_list)):
    #adding all numbers of list
    sum += numbers_list[i]
  #returning sum of the list as per assignment relquirement
  return sum

#call the function by passing list as a parameter
calculate_sum([1, 2, 3, 4, 5])
```

```python
"""Write a function called remove_duplicates that takes a list as an argument and
returns a new list with duplicate elements removed."""

#defining function to remove duplicate of a given list, method=1
def remove_duplicates(numbers_list):
  #converting list to set datatype as set don't have any duplicate
  new_list = set((numbers_list))
  #returning sets as a list as per assignment relquirement
  return list(new_list)

#call the function by passing list as a parameter
remove_duplicates([1, 2, 2, 3, 4, 4, 5])
```

```python
"""Write a function called find_max that takes a list of numbers as an argument and
returns the largest number in the list."""

#defining a function which will return the largest number in a list
def find_max(numbers_list):
  #using built-in python function to find largest entry
  largest_number = max(numbers_list)
  #returning largest number as per assigment requirement
  return largest_number

#call the function by passing list as a parameter
find_max([10, 5, 8, 20, 3])
```

```python
"""Write a function called remove_duplicates that takes a list as an argument and
returns a new list with duplicate elements removed."""

#defining function to remove duplicate of a given list, method=2
def remove_duplicates(numbers_list):
  #defining a variable of list datatype
  new_list = []
  #iterating for loop for all numbers in given list
  for i in range(len(numbers_list)):
    #remove first entry of given list
    current_number = numbers_list.pop(0)
    #check if removed entry is not present in given list
    if current_number not in numbers_list:
      #if true, add this entry to new list
      new_list.append(current_number)
  #returning new list as per assigment requirement
  return new_list

#call the function by passing list as a parameter
remove_duplicates([1, 2, 2, 3, 4, 4, 5])
```

```python
""" Write a Function called Tweet_generator() that takes input string from the user
and return the Tweet generated on the given topic. Use the open.ai API for this
task."""

#import OpenAi libraries
import openai

#unique openai key for authentication
openai.api_key = "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"

#define a function to generate tweet on a given topic
def Tweet_generator(text):
    #defining a variable named message to interact with chatgpt
    message = [
        {"role": "system", "content": """You are trained to analyse a given topic and generate tweet.
        Tweet must contains at least 50 words, also add 3 related hashtags at the end."""},
        {"role": "user", "content": f"""Analyse the following topic and generate a tweet. The topic is {text}
        The tweet must contain the following format. First text of tweet and then 3 hashtags"""}
    ]

    #initializing chatgpt to complete chat with parameters like model, tokens and temprature
    response = openai.ChatCompletion.create(
        model = "gpt-3.5-turbo",
        messages = message,
        max_tokens = 60,
        n = 1,
        stop = None,
        temperature = 0.5
    )

    #extracting content from JSON script(given at the last for reference)
    tweet = response.choices[0].message.content.strip()
    return tweet

#calling function and printing result as per assigment requirement
Tweet_generator(input("Input text to generate tweet: "))
```
