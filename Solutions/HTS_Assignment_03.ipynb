```python
# Write a program that prints the multiplication table of a given number using a for loop

#define a number for multiplication table
number = 2
#change to integat datatype
n = int(number)

#iterating for loop from 1 to 10 and printing result for the table
for i in range(1,11):
  print(f"{n} x {i} = {i*n}")
```

```python
# Write a program that prints all the even numbers from 1 to 10 using a while loop


#initiating i to iterate while loop
i = 1
#iterating wile loop from 1 to 10
while i <= 10:
  #print number if quotient is equal to zero
  if i%2 == 0:
    print(i)
  #increment i for next iteration
  i += 1
```

```python
# Write a function called "is_even" that takes a number as an argument and returns True if the number is even, and False otherwise.

#define funnction which will print True if quotient is equal to zero, else it will print False
def is_even(number):
  n = int(number)
  if n%2 == 0:
    print(True)
  else:
    print(False)

#prompting user for a number
is_even(input("Type a positive integar: "))
```
