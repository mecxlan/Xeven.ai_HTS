#Task 2: Handling Missing Values

```python
# 1. Check the dimension of dataset using the .shape() function.

df.shape
```

```python
# 2. Determine the percentage of missing values for each feature.

for column in df.columns:
  missing_percent = (df[column].isnull().sum() * 100) / len(df)
  print(f"{column}   {missing_percent:.1f}")
```

```python
# 3. Remove the missing values from the dataset.

df = df.dropna()
df.count()
```

```python
# 4. Again, check the number of missing values in dataset.

for column in df.columns:
  missing_percent = (df[column].isnull().sum() * 100) / len(df)
  print(f"{column}   {missing_percent:.1f}")
```

```python
# 5. Check for duplicate values in the dataset and display the number of duplicate rows

duplicate_rows = df[df.duplicated()]
print(f"Number of duplicate rows are {duplicate_rows.shape[0]}")
```

```python
# 6. Remove the duplicate rows from the dataset

df = df.drop_duplicates()
```

```python
# 7. Again, display the number of rows of data frame to verify the duplicate values are
# removed

df.count()
```

#Task 3: Descriptive Statistics

```python
# 1. Calculate basic summary statistics (mean, median, min, max, etc.) for the numerical
# features. Using .describe()

df.describe()
```

```python
# 2. Calculate the basic summary statistics (mean, median, min, max, etc.) for the numerical
# features. Using .quantile()

df.quantile([0, 0.25, 0.5, 0.75, 1], numeric_only="True")
```

```python
# 3. Calculate the correlation matrix for all numerical features.

correlation = df.corr(numeric_only="True")
correlation
```

```python
# 4. Visualize the correlation matrix using a heatmap.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

sns.heatmap(correlation, cmap = "BrBG", annot = True)
```
